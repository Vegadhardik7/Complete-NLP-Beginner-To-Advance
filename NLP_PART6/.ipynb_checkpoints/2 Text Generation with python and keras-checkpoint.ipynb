```python
import pandas as pd
import numpy as np
import spacy

nlp = spacy.load('en_core_web_lg')
```

creating ***seperate_pun*** function which take in some doc as string and then its going to grab the text tokens. If they're not some particular type of <b>Punctuation</b> or they are not a <b>New Line</b>. 

```python
def seperate_pun(doc_text):
    return [token.text.lower() for token in nlp(doc_text) 
            if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

This means:
    
1.) Grab moby_dick_four_chapters.txt ----------------------------> doc_text

2.) Use nlp on it -----------------------------------------------------> nlp(doc_text)

3.) Than we iterate it through each token --------------------------------> token

4.) Grab text and make them into lowercase -----------------------------------> token.text.lower()

#### How ever we wanna get rid of things that are probably not going to be very helpful for training purposes like . or \n because they show up so often in the actual text espically new lines. I want to make sure my text generation neural network dosent overfit that sort of punctuation.

So solution is:

token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n '

# Part 2:

### * Creating LSTM Based Model
### * Splitting Data into Features and Labels
#### - X Feature (First n words of Sequence)
#### - Y Label (Next word after the sequence)
### * Fit the model

```python
from keras.utils import to_categorical

sequences[:,:-1] # Grab evey thing except that last word
```

```python
# last word in the seq will be the last column

sequences[:,-1]
```

```python
X = sequences[:,:-1]
y = sequences[:,-1]
```

```python
y = to_categorical(y, num_classes = vocabulary_size+1)   # +1 bcz it need extra 1 to hold the zero

seq_len = X.shape[1]
X.shape
```

```python
# creating the model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding

def create_model(vocabulary_size, seq_len):
    
    model = Sequential()
    
    model.add(Embedding(vocabulary_size, seq_len, input_length=seq_len))
    
    model.add(LSTM(50,return_sequences=True))
    model.add(LSTM(50))
    model.add(Dense(50,activation='relu'))
    
    model.add(Dense(vocabulary_size, activation='softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1, seq_len)
```

```python
from pickle import dump, load

# Feature label split
model.fit(X,y,batch_size=128,epochs=200,verbose=1)
```

```python
model.save('My_mobydick_file.h5')

dump(tokenizer, open('my_simple_tokenizer', 'wb'))
```

### Generate New Text Based on Seed Input

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

def generate_text(model,tokenizer,seq_len,seed_text,num_word_gen):
    
    output_text = []
    
    # 25 words
    input_text = seed_text
    
    for i in range(num_word_gen):
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len,truncating='pre')
        
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        
        pred_word = tokenizer.index_word[pred_word_ind]
        
        input_text = ' '+pred_word
        
        output_text.append(pred_word)
        
    return ' '.join(output_text)
```
