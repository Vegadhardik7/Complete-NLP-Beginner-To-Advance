```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_path = "/content/drive/MyDrive/Python_Books/Corona_NLP_train.csv"
test_path = "/content/drive/MyDrive/Python_Books/Corona_NLP_test.csv"
```

```python
df_train_path = pd.read_csv(train_path, encoding='latin1')
df_test_path = pd.read_csv(test_path, encoding='latin1')
```

```python
df = df_train_path.append(df_test_path, ignore_index=True)
df.head()
```

```python
df = df.iloc[:,-2:]
df.head()
```

```python
df.Sentiment = df.Sentiment.replace('Extremely Positive', 'Positive')
df.Sentiment = df.Sentiment.replace('Extremely Negative', 'Negative')


df.Sentiment = df['Sentiment'].map({'Neutral':0, 'Positive':1, 'Negative':2})
```

```python
df.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
X = df.drop('Sentiment',axis=1)
y = df['Sentiment']

messages=X.copy()

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub(r'http\S+', '', messages['OriginalTweet'][i])
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot

voc_size=10000

onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
num_classes = 3

# Number of dimenstion for word embedding
embed_num_dims = 300

# Max input length (max num of words)
max_seq_len = 500
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
# sent_length=100
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=max_seq_len)
print(embedded_docs)
```

```python
embedded_docs[672]
```

```python
df['Sentiment'].head()
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y)
y
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)

X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_final,y_final,test_size=0.2)
```

## Bidirectional Multilayer GRU

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, Embedding, GRU
from tensorflow.keras.callbacks import EarlyStopping

embedding_vector_features=128

# embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=max_seq_len))
model.add(Dropout(0.5))

model.add(Bidirectional(GRU(100,return_sequences=True)))
model.add(Dropout(0.5))

model.add(Bidirectional(GRU(50,return_sequences=True)))
model.add(Dropout(0.5))

model.add(Bidirectional(GRU(10)))
model.add(Dropout(0.5))

model.add(Dense(3,activation='softmax'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1, mode='min')

model.fit(X_train, y_train,epochs=50,batch_size=64,validation_data=(X_test, y_test),callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)

plt.plot(loss)
```

```python
# Model Evaluation

class_names = [0,1,2]

pred_gru = model.predict(X_test) 
pred_gru = np.argmax(pred_gru, axis=1)
pred_gru = [class_names[pred] for pred in pred_gru]
```

```python
print('Message: {}\nPredicted: {}'.format(X_test[4], pred_gru[4]))
```
