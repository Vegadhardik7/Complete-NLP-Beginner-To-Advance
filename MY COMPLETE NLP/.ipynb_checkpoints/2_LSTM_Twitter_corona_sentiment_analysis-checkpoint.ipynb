```python
df = pd.concat((df_train,df_test))

df.reset_index(inplace=True)

df.head()
```

```python
df = df.drop('index', axis=1)
df
```

```python
df['Sentiment'].value_counts()
```

```python
# df['Sentiment'] = df['Sentiment'].map({'Extremely Positive': 'Positive', 'Extremely Negative': 'Negative'})

df.Sentiment = df.Sentiment.replace('Extremely Positive', 'Positive')
df.Sentiment = df.Sentiment.replace('Extremely Negative', 'Negative')


df.Sentiment = df['Sentiment'].map({'Neutral':0, 'Positive':1, 'Negative':2})
```

```python
import nltk
import re
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
# data = df.OriginalTweet
# data.replace("[^a-zA-Z#@]"," ",regex=True,inplace=True)
# data = data.apply(lambda x: x.astype(str).str.lower())
from textblob import TextBlob

X = df.drop('Sentiment',axis=1)
y = df['Sentiment']


messages=X.copy()
messages.reset_index(inplace=True)


from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub(r'http\S+', '', messages['OriginalTweet'][i])
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
len(corpus)
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot

voc_size=10000

onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
X.shape
```

```python
y.shape
```

```python
# from sklearn.feature_extraction.text import CountVectorizer

# cv = CountVectorizer(ngram_range=(2,2))
# val = cv.fit_transform(corpus).toarray()
```

```python
# val
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
sent_length=50
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)

X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 3)
y_test = to_categorical(y_test, 3)
```

# MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, Embedding, GRU
from tensorflow.keras.callbacks import EarlyStopping

embedding_vector_features=128

model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))

model.add(GRU(100))
model.add(Dropout(0.3))

# model.add(GRU(50))
# model.add(Dropout(0.3))

model.add(Dense(3,activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=1, mode='min')

model.fit(X_train, y_train,epochs=50,batch_size=64,validation_data=(X_test, y_test),callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(loss)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, Embedding, GRU
from tensorflow.keras.callbacks import EarlyStopping

embedding_vector_features=128

models=Sequential()
models.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
models.add(Dropout(0.3))

models.add(GRU(100,return_sequences=True))
models.add(Dropout(0.3))

models.add(GRU(50,return_sequences=True))
models.add(Dropout(0.3))

models.add(Dense(3,activation='softmax'))

models.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(models.summary())
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=1, mode='min')

models.fit(X_train, y_train,epochs=50,batch_size=64,validation_data=(X_test, y_test),callbacks=[early_stop])
```

```python
loss = pd.DataFrame(models.history.history)

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(loss)
```

```python
from keras.layers import Dense, Embedding, LSTM, SpatialDropout1D
from keras.models import Sequential
from sklearn.feature_extraction.text import CountVectorizer
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical
from keras.callbacks import EarlyStopping

epochs = 50
# emb_dim = 128
batch_size = 256



voc_size=10000
sent_length=20
embedding_vector_features=128

model2 = Sequential()
model2.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model2.add(SpatialDropout1D(0.5))
model2.add(LSTM(64, dropout=0.5, recurrent_dropout=0.5))
model2.add(Dense(3, activation='softmax'))
model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
print(model2.summary())
history = model2.fit(X_train, y_train, epochs=epochs, batch_size=batch_size,validation_split=0.2,callbacks=[EarlyStopping(monitor='val_loss',patience=7, min_delta=0.0001)])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout, Embedding, GRU
from tensorflow.keras.callbacks import EarlyStopping

embedding_vector_features=128

# embedding_vector_features=40
model3=Sequential()
model3.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model3.add(Dropout(0.3))

model3.add(Bidirectional(LSTM(100,return_sequences=True)))
model3.add(Dropout(0.3))

model3.add(Bidirectional(LSTM(50,return_sequences=True)))
model3.add(Dropout(0.3))

model3.add(Dense(3,activation='softmax'))
model3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model3.summary())
```

```python
early_stop1 = EarlyStopping(monitor='val_loss', patience=20, verbose=1, mode='min')

model3.fit(X_train, y_train,epochs=50,batch_size=64,validation_data=(X_test, y_test),callbacks=[early_stop1])
```

```python

loss = pd.DataFrame(model3.history.history)

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(loss)
```
