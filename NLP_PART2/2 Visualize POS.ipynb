```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(u"The quick brown fox jumped over a lazy dog's back.")
print(doc)
```

```python
from spacy import displacy

displacy.render(doc,style='dep',jupyter=True, options={'distance':110})
```

```python
options = {'distance':110, 'compact': 'True', 'color': 'red', 'bg': '#FFFF00', 'font':'Times'}

displacy.render(doc,style='dep',jupyter=True, options=options)
```

### If we have a larger text we have to pass span of text

```python
doc2 = nlp(u"This is a sentence. This is another sentence, possibly longer than the other.")
```

```python
spans = list(doc2.sents)
```

```python
displacy.serve(spans, style='dep', options = {'distance':110})
```
